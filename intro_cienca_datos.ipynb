Ahora,se limpia el dataframe de "impurezas" Primero  "Purchase_Amount"I wu numerico, que tiene 13 valores NaN. se peuden reemplazar los NaN po un numero atipico, por ejemplo 999
luego, se reemplazan los 999 por la media o la moda de la columna

```python
df["Purchase_Amount"] = df["Purchase_Amount"].replace(np.nan, 999)
```

```python
df
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].abs()
```

```python
df
```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].replace(999, df["Purchase_Amount"].mean())
df["Purchase_Amount"] = df["Purchase_Amount"].round(2)
```

```python
df
```

```python
df['Store_Location'] = df['Store_Location'].str.upper()
df
```

```python
df["Product_Category"]=df["Product_Category"].str.upper()
df
```

```python
df["Quantity"] = df["Quantity"].replace("two", 2)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors='coerce')
df["Quantity"] = df["Quantity"].replace(np.nan, df["Quantity"].mean())
df["Quantity"] = df["Quantity"].round(2)
df
```

```python
df["Email"]=df["Email"].str.lower()
df["Email"] = df["Email"].fillna(df["Name"].str.lower().str.replace(' ', '.') + "@hotmail.com")
df
```

```python
df["Name"] = df["Name"].str.title()
df
```

```python
df.to_csv("clean_dataset.csv", index=True)
df.head()
```
