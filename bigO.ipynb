# **Gu√≠a Completa sobre la Notaci√≥n Big O**

La **notaci√≥n Big O** es una herramienta fundamental en la inform√°tica y la ciencia de datos. Nos permite medir el rendimiento de los algoritmos en funci√≥n del tama√±o de la entrada (*n*), enfoc√°ndonos en c√≥mo crece el tiempo de ejecuci√≥n o el uso de memoria cuando los datos aumentan.

## **¬øPor qu√© es importante la Notaci√≥n Big O?**

Cuando dise√±amos o elegimos un algoritmo, queremos que sea lo m√°s eficiente posible. Big O nos ayuda a comparar diferentes soluciones y entender cu√°l se comportar√° mejor a medida que trabajemos con m√°s datos.

---

## **Principales Clases de Complejidad**

A continuaci√≥n, te presentamos las complejidades m√°s comunes en los algoritmos:

| Notaci√≥n | Nombre | Ejemplo |
|----------|--------|---------|
| **O(1)** | Constante | Acceder a un elemento de un array por √≠ndice |
| **O(log n)** | Logar√≠tmica | B√∫squeda binaria |
| **O(n)** | Lineal | B√∫squeda secuencial |
| **O(n log n)** | Cuasilineal | Ordenamiento r√°pido (QuickSort) |
| **O(n¬≤)** | Cuadr√°tica | Ordenamiento burbuja (Bubble Sort) |
| **O(2‚Åø)** | Exponencial | Algoritmos de fuerza bruta en problemas NP |

---

## **Ejemplos con C√≥digo**

### **Ejemplo de O(1) ‚Äì Tiempo Constante**
Este c√≥digo siempre toma el mismo tiempo, sin importar el tama√±o de la lista.  

```python
def obtener_primero(lista):
    return lista[0]  # Siempre accede al primer elemento
```
üìå **Independientemente del tama√±o de `lista`, solo hace una operaci√≥n.**

---

### **Ejemplo de O(n) ‚Äì Tiempo Lineal**
Recorrer una lista es una operaci√≥n de **O(n)**, ya que aumenta proporcionalmente con `n`.

```python
def buscar_elemento(lista, objetivo):
    for elemento in lista:
        if elemento == objetivo:
            return True
    return False
```
üìå **Si `n` es grande, el tiempo de ejecuci√≥n tambi√©n lo ser√°.**

---

### **Ejemplo de O(n¬≤) ‚Äì Tiempo Cuadr√°tico**
Este c√≥digo usa un doble bucle anidado, lo que hace que su tiempo de ejecuci√≥n crezca cuadr√°ticamente.

```python
def pares_unicos(lista):
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            print(lista[i], lista[j])  # Comparaci√≥n de cada elemento con todos los dem√°s
```
üìå **Si `n = 1000`, las operaciones ser√°n alrededor de `1000¬≤ = 1,000,000`.**

---

## **Diagrama de Comparaci√≥n de Complejidades**

üìà **Gr√°fico de crecimiento de Big O**





````python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de n
n = np.linspace(1, 10, 100)

# Definir las funciones de complejidad
O_1 = np.ones_like(n)                # O(1) - constante
O_log_n = np.log(n)                  # O(log n) - logar√≠tmica
O_n = n                               # O(n) - lineal
O_n_log_n = n * np.log(n)            # O(n log n) - cuasilineal
O_n2 = n**2                          # O(n^2) - cuadr√°tica
O_2n = 2**n                          # O(2^n) - exponencial

# Graficar las funciones
plt.figure(figsize=(10, 6))
plt.plot(n, O_1, label="O(1)", linewidth=2)
plt.plot(n, O_log_n, label="O(log n)", linewidth=2)
plt.plot(n, O_n, label="O(n)", linewidth=2)
plt.plot(n, O_n_log_n, label="O(n log n)", linewidth=2)
plt.plot(n, O_n2, label="O(n¬≤)", linewidth=2)
plt.plot(n, O_2n, label="O(2‚Åø)", linewidth=2, linestyle="dashed")

# Configuraci√≥n del gr√°fico
plt.ylim(0, 100)  # Limitar el eje Y para mejor visualizaci√≥n
plt.xlabel("Tama√±o de Entrada (n)")
plt.ylabel("Operaciones")
plt.title("Crecimiento de Complejidades en Notaci√≥n Big O")
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.show()
````


üí° **Cuanto m√°s r√°pido crece la funci√≥n, menos eficiente es el algoritmo en datos grandes.**

Si necesitas comparar algoritmos para un problema, Big O te ayuda a elegir el m√°s eficiente.
